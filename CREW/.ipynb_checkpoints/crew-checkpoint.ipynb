## W8 where = sec/sub/sys = pos

```python
sec="Spinward Marches"
sub="C Regina"
sys="1910 Regina"

loc = sec + "/" + sub + "/"  + sys
print(loc)

world = sub[2:]
print(world)

pos = sys[0:4]
print(f"{pos}")

uwp = "A788899-C"
sp = uwp[0:1]
si = uwp[1:2]
at = uwp[3:4]
hy = uwp[2:3]
po = uwp[4:5]
go = uwp[5:6]
ll = uwp[6:7]
tl = uwp[8:9]
uwps = sp + si + at + hy + po + go + ll + "-" + tl
assert uwp == uwps
print(loc + " " + world + " = " + uwps)
```
